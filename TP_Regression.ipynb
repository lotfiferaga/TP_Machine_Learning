```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 200
x = np.random.rand(n)
```

```python
y = np.random.rand(n)
```

```python

plt.scatter(x,y)
```

```python
un= np.ones(200).T
```

```python
newX= np.column_stack((un, x.T))
```

```python
teta = y.T @ newX @ np.power(newX.T @ newX, -1)
```

```python
teta-1
```

```python
plt.plot(teta-1,"r")
plt.scatter(x,y)
```

```python
alpha = 1
teta2  = np.dot(np.dot(y, newX), np.power(np.dot(newX.T,newX)+ alpha * np.identity(2),-1))
```

```python
plt.plot(teta2-1, "r")
plt.scatter(x,y)
```
